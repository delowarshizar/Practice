```python
topics = {
    'pyhton Basics': {'difficulty': 'beginner','des': 'BP'},
    'Data_stucture': {'difficulty': 'intermidiate','des': 'data_stucture'},
    'advanced python': {'difficulty': 'adv','des': 'adv_pt'},
    'machine_learning': {'difficulty': 'expert','des' : 'intro_ml'},
}

for node, attr in topics.items():
    G.nodes[node].update(attr)

edge_attributes = {
      ('pyhton Basics','Data_stucture'):{'weight':1.0,'prerequisite': True},
      ('Data_stucture','advanced python'):{'weight':1.5,'prerequisite': True},
      ('advanced python','machine_learning'):{'weight':2.0,'prerequisite': True},
}

for edge, attr in edge_attributes.items():
    G.edges[edge].update(attr)

pos = nx.spring_layout(G)
nx.draw(G, pos,with_labels=True)

node_labels = {node: data['difficulty'] for node, data in G.nodes(data=True)}
for node,(x,y) in pos.items():
    plt.text(x,y-0.05, node_labels[node], fontsize=8, ha = 'center', color = 'darkgreen' )
    edge_labels = {(u,v): f"weight: {d['weight']}, prereq: {d['prerequisite']}" for u,v,d in G.edges(data=True)}

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size= 8)
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load CSV
df = pd.read_csv("/content/drive/MyDrive/Dataset/indian-cities-dataset.csv")

# Use only the first 10 rows
df_10 = df.iloc[:10]

# Create an undirected graph
G = nx.Graph()

# Add edges and distances
for _, row in df_10.iterrows():
    origin = row['Origin']
    destination = row['Destination']
    distance = row['Distance']
    G.add_edge(origin, destination, weight=distance, label=distance)

# Choose BFS source node (first origin city)
source = df_10.iloc[0]['Origin']

# Perform BFS traversal
bfs_edges = list(nx.bfs_edges(G, source=source))
bfs_nodes = [source] + [v for u, v in bfs_edges]

# Assign a unique color to each node
nodes = list(G.nodes())
num_nodes = len(nodes)
cmap = cm.get_cmap('tab20', num_nodes)  # 'tab20' has 20 distinct colors
node_colors = [cmap(i) for i in range(num_nodes)]
color_map_dict = dict(zip(nodes, node_colors))
node_color_list = [color_map_dict[node] for node in nodes]

# Layout using spring layout with distance weighting
pos = nx.spring_layout(G, weight='weight', k=0.5, iterations=100)

# Draw graph
plt.figure(figsize=(12, 8))
nx.draw_networkx_nodes(G, pos, node_color=node_color_list, node_size=800)
nx.draw_networkx_edges(G, pos, width=2, alpha=0.6)
nx.draw_networkx_labels(G, pos, font_size=10)

# Draw edge labels (distance)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

plt.title(f"BFS Traversal on First 10 Rows (Source: {source})", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
